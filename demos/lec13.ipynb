```python
import seaborn as sns
import pandas as pd
import random
```

# Randomly select from a set of choices

`random.choice`

`random.choices`

`random.sample`

```python
random.choice(['H','T']) # By default uniformly distribution
```

```python
random.choices(['H','T'],k=10) # choice vs. choiceS
```

```python
sns.histplot(random.choices(['H','T'],k=10000))
```

```python
random.choices(['H','T'], weights=[0.99,0.01], k=10)
```

```python
sns.histplot(random.choices(['H','T'], weights=[0.99,0.01], k=10000))
```

```python
random.choices(range(1,40), k=20) # with replacement
```

```python
random.sample(range(1,40), k=20) # without replacement
```

# Sample rows from table

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sample.html

```python
penguins = pd.read_csv('palmer_penguins.csv')
penguins.head() # top 5
```

```python
penguins.sample() # random 1
```

```python
penguins.sample(5) # random 5
```

```python
penguins.sample(5, ignore_index=True) # random 5 then relabel index
```

```python
penguins.sample(frac=0.01, ignore_index=True) # random 1% then relabel index
```

```python
penguins.sample(5, random_state=13) 
# if you want reproducible results. random_state can be any integer
```

```python
penguins.sample(5, random_state=13, replace=True) # sample with replacement
```
