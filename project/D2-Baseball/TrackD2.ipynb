# Merging (or joining) tables

- Intuitive SQL explanations on `left`, `right`, `inner`, `outer` join of tables: https://www.geeksforgeeks.org/sql/sql-join-set-1-inner-left-right-and-full-joins/

- Long but thorough tutorial: https://realpython.com/pandas-merge-join-and-concat/


```python
# add salary information (from salary table) to each player in hitting table
players_merged = pd.merge(hitting, salary, how='left', on=['playerID', 'yearID', 'teamID', 'lgID'])
players_merged
```

You can join any two tables. You can also do multiple of them by saving the merged table to the same variable.
